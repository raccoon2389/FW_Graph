```python
import pandas as pd
```

```python
df = pd.read_csv('./04_hashed.csv')
```

```python
src = df[['src_ip','dst_ip']].groupby('src_ip').nunique().sort_values('dst_ip',ascending=False)
```

```python
dst = df[['src_ip','dst_ip']].groupby('dst_ip').nunique().sort_values('src_ip',ascending=False)
```

```python
src.columns=['count']
```

```python
dst.columns=['count']
```

```python
count = src.add(dst,fill_value=0).sort_values('count',ascending=False)
```

```python
mask = [ i for i in range(len(count))]
count['mask'] = mask
```

```python
count['is_interner'] = count['count']>count['mask']
```

```python
def ppap(x):
    if x in flag:
        return 'outbound'
    else:
        return 'inbound'
        
```
